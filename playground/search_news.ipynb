```python
import asyncio
import json

from src.tools.search import search_news
from src.tools.sources import list_sources
```

```python
async def run_list_sources(**kwargs):
    raw = await list_sources(**kwargs)
    return json.loads(raw)


async def run_search(**kwargs):
    raw = await search_news(**kwargs)
    data = json.loads(raw)
    if not data.get("ok"):
        return data

    return {
        "ok": True,
        "count": data.get("count"),
        "articles": [
            {
                "name": article.get("name"),
                "title": article.get("title"),
                "tier": article.get("tier"),
                "category": article.get("category"),
                "published_at": article.get("published_at"),
                "url": article.get("url"),
            }
            for article in data.get("articles", [])
        ],
    }


async def demo():
    return {
        "sources": await run_list_sources(tiers=[1], limit=10),
        "news": await run_search(sources=["Federal Reserve"], timespan="30d", limit=5),
    }


await demo()
```
